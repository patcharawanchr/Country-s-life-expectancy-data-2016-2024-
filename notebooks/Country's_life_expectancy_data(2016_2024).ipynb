```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sudhakumari1/countrys-life-expectancy-data2016-2024")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os

# List the contents of the directory where the dataset was downloaded
download_path = "/kaggle/input/countrys-life-expectancy-data2016-2024"
print(os.listdir(download_path))
```

```python
df=pd.read_csv("/kaggle/input/countrys-life-expectancy-data2016-2024/Life Expactancy and GDP (2016-2024).csv", encoding='latin-1')
print('Shape',df.shape)
df.head()
```

```python
# Missing values
print(df.isna().sum())

# Duplicates
print("Duplicates:", df.duplicated().sum())

# ชนิดข้อมูล
print(df.dtypes)

# Descriptive statistics
df.describe()
```

```python
# Distribution ของ Life Expectancy
df["Life Expectancy"].hist(bins=30)
plt.title("Distribution of Life Expectancy")
plt.xlabel("Years")
plt.ylabel("Count")
plt.show()

# ค่าเฉลี่ยการศึกษา (Schooling)
print("Average Schooling:", df["Schooling"].mean())
```

```python
# เปรียบเทียบ Life Expectancy ระหว่าง Developed vs Developing
sns.boxplot(x="Status", y="Life Expectancy", data=df)
plt.title("Life Expectancy by Status")
plt.show()

# ความสัมพันธ์ GDP vs Life Expectancy
sns.scatterplot(x="GDP", y="Life Expectancy", hue="Status", data=df, alpha=0.7)
plt.title("GDP vs Life Expectancy")
plt.show()
```

```python
num_df = df.select_dtypes(include="number")
plt.figure(figsize=(10,7))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Convert 'GDP' to numeric, coercing errors
df["GDP"] = pd.to_numeric(df["GDP"], errors='coerce')

# แบ่งกลุ่ม GDP (Low / Middle / High)
df["GDP_Group"] = pd.qcut(df["GDP"], q=3, labels=["Low","Medium","High"])

# ค่าเฉลี่ยอายุขัยในแต่ละกลุ่ม
print(df.groupby("GDP_Group")["Life Expectancy"].mean())
```

```python
# Export cleaned data
df.to_csv("cleaned_data.csv", index=False)

# Save รูป
plt.savefig("life_expectancy_distribution.png")
```
